# **Data Preparation**

To start Data Preparation, we start with data cleaning. And to start data cleaning, we first get the subset of the dataset. that means, we take the columns which are required.


```python
df.columns
```

```python
# let's just take the column which we need
df = df[['coaster_name',
         #'Length', 'Speed',
         'Location', 'Status',
         'Opening date',
         #'Type',
         #'Manufacturer', 'Height restriction', 'Model', 'Height',
         #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
         #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
         #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
         #'Track layout', 'Fastrack available', 'Soft opening date.1',
         #'Closing date', 'Opened', 'Replaced by', 'Website',
         #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
         #'Single rider line available', 'Restraint Style',
         #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
         'year_introduced', 'latitude', 'longitude', 'Type_Main',
         'opening_date_clean',
         #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
         'speed_mph',
         #'height_value', 'height_unit',
         'height_ft',
         'Inversions_clean', 'Gforce_clean']].copy()

#.copy() just creates a brand new copy of object which is independent of original copy
```

So in order to get the subset, there is one more method called .drop().

```python
df.drop(['Opening date'], axis=1) #axis in order to define the row or column
```

```python
df.dtypes
```

```python
#If we have to change the datatype of some values
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean']) 
```

Renaming Columns

```python
df.columns
```

```python
#renaming the column names
df = df.rename( columns = {'coaster_name':'Coaster_Name',
               'year_introduced':'Year_Introduced',
                           'opening_date_clean':'Opening_Date',
                           'speed_mph':'Speed_mph',
                           'height_ft':'Height_ft',
                           'Inversions_clean':'Invensions',
                           'Gforce_clean':'Gforce'
                })
```

```python
df.head()
```

```python
#df.isna() #gives the statistical display of null values
df.isna().sum() #gives the sum of count of total null values in column
```

```python
#To find that if there is duplicate values in fields
df.duplicated()
```

```python
# to find out which one is duplicated, we can locate by;
df.loc[df.duplicated()]
```

```python
# so in a case, we want to locate duplicated in subset like particular columns then
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
#we can also query on where exactly is duplications by;
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
#Now we find the duplications, we might want to delete it so
#df.columns
~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])
#Here what tilde(~) does is, inverse the duplicated function.suppose when we want those columns which are not duplicate
#And to locate them ;
df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])]
```

```python
#Now that we have done some works at columns, indexing might have been affected / might have been jumped off
#In order to reset that;
df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
 .reset_index(drop=True)
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
 .reset_index(drop=True).copy()
```

```python
# So In order to understand the feature distribution, we have some functions like ;
df['Year_Introduced'].value_counts()
# This gives the occurance of the value in descending order
```

```python
#plotting the count whose occurance is highest
ax = df['Year_Introduced'].value_counts() \
.head(10) \
.plot(kind='bar', title = 'Top Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
# next thing we can do to understand the feature is it's distribution
df['Speed_mph'].plot(kind='hist', bins=20, title='Coaster Speed (mph)')
#Histogram
```

```python
#density plot
df['Speed_mph'].plot(kind='kde',title='Coaster Speed (mph)')
```
